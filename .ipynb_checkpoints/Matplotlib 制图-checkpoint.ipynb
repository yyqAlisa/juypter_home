# figure的使用

```python
data = np.arange(-3,3,0.1)
x = np.sin(data)
y = np.cos(data)

#画图
plt.figure()
plt.plot(data,x) #第一个线图

plt.figure(figsize=(8,6))
plt.plot(data,y) #第二个线图
plt.show()
```

# 刻度、标题、标签和图例

```python
x = [1,3,5]
y = [2,4,3]

#创建图片
plt.figure()
#画图
plt.plot(x,y,'r--',label = 'x-line')
#生成图例
plt.legend(loc = 'upper left')

# #设置标题
plt.title('进球数')

# #设置x/y轴标签
plt.xlabel('x轴')
plt.ylabel('y轴')

# #设置刻度的范围
# plt.xlim(1,10)
# plt.ylim(1,10)

#设置图标的刻度
plt.xticks(np.arange(1,6,1),[str(i)+'月' for i in range(1,6,1)])
# plt.xticks(np.arange(1,6,1),['中国','2月','3月','4月','5月'])          
plt.yticks(np.arange(1,6,1))
           
# #获取坐标轴信息
ax = plt.gca()
# ax.spines['right'],set_color(None)
# ax.spines['top'],set_color(None)
ax.spines['top'].set_visible(False) #去掉上边框
# ax.spines['bottom'].set_visible(False) #去掉下边框
# ax.spines['left'].set_visible(False) #去掉左边框
ax.spines['right'].set_visible(False) #去掉右边框

#展示图片
plt.show()
```

# 子图面向对象的写法

```python
#创建图形
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax1.plot(np.random.randn(50).cumsum(),'r--')

ax2 = fig.add_subplot(222)
ax2.plot(np.random.randn(50).cumsum(),'b-')

ax3 = fig.add_subplot(223)
ax3.plot(np.random.randn(50).cumsum(),'y--')

ax4 = fig.add_subplot(224)
ax4.plot(np.random.randn(50).cumsum(),'g-')

plt.show()
```

# 柱状图

```python
#设置数据
height1 = [100,200,300,400] # y  #要改
height2 = [120,220,320,420]
height3 = [110,210,310,410]

#设置数据标量
left = np.arange(1,5,1) # 4 x   #要改

#设置柱子宽度
n = 4
width = 0.8/4

#创建figure
plt.figure()

#画图
plt.bar(left,height1,width = width,color = 'lightskyblue',align = 'center',label = '上海')  #align 柱子的位置默认在center
plt.bar(left-0.2,height2,width = width,color = 'red',align = 'center',label = '北京')
plt.bar(left+0.2,height2,width = width,color = 'green',align = 'center',label = '深圳')
    
#生成图例
plt.legend(loc = 'upper left')

# #设置标题
plt.title('旅游人数')     #要改

# #设置x/y轴标签
plt.xlabel('月份')        #要改
plt.ylabel('人次/百万')

#设置图标的刻度
plt.xticks(np.arange(1,5,1),[str(i)+'月' for i in range(1,5,1)])    #要改
# plt.xticks(np.arange(1,5,1),['1月','2月','3月','4月')          
plt.yticks(np.arange(100,600,100))
           
# #获取坐标轴信息
ax = plt.gca()
ax.spines['top'].set_visible(False) #去掉上边框
ax.spines['right'].set_visible(False) #去掉右边框

#展示图片
plt.show()
```

# 直方图

```python
#设置数据
mu,sigma = 100,15
x = np.random.randn(1000)*mu*sigma
x = np.random.randn(1000).cumsum()

plt.figure()
plt.hist(x,100,alpha = 0.5,color = 'red')   #100表示多少根柱子，alpha 透明度

plt.title('频率图')

plt.grid(True)   #加网状
plt.show()
```

# 散点图

```python
#设置数据
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.scatter(x,y,c = np.random.randn(1000),s = np.random.randn(1000)*50,alpha = 0.7)

plt.title('aaa')

plt.show()
```

# 饼图

```python
#设置数据
x = [15,20,25,30]

#设置颜色
color = ['red','green','blue','yellow']

#设置每一块离开中心的距离
explode = [0.1,0,0,0]

#每一块的文字说明
labels = ['娱乐','日常生活','女朋友','游戏']

plt.figure()

#绘制饼图
plt.pie(x,explode = explode,labels = labels, colors = color,startangle = 90,shadow = True,labeldistance = 1.1,autopct ='%.1f%%')

#设置圆图
plt.axis('equal')

#设置标题
plt.title('月工资分配')

#展示图例
plt.show()
```
